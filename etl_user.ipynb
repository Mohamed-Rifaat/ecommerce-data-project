```python
import random
from datetime import datetime, timedelta

orders = []

for i in range(1, 201):
    user_id = random.randint(1, 100)
    
    random_days = random.randint(0, 30)
    order_date = datetime.now() - timedelta(days=random_days)
    
    orders.append((
        i,
        user_id,
        order_date.strftime("%Y-%m-%d")
    ))

print("Orders data generated!")
```

```python
cursor.executemany(
    "INSERT INTO dbo.orders (order_id, user_id, order_date) VALUES (?, ?, ?)",
    orders
)

conn.commit()

print("Orders inserted successfully!")
```

```python
order_items = []

order_item_id = 1

for order in orders:
    order_id = order[0]
    
    # كل أوردر فيه من 1 لـ 3 منتجات
    num_items = random.randint(1, 3)
    
    for _ in range(num_items):
        product_id = random.randint(1, 10)
        quantity = random.randint(1, 5)
        
        # سعر عشوائي (تقريبًا من اللي عندنا)
        unit_price = random.randint(100, 15000)
        
        order_items.append((
            order_item_id,
            order_id,
            product_id,
            quantity,
            unit_price
        ))
        
        order_item_id += 1

print("Order items generated!")
```

```python
cursor.executemany(
    "INSERT INTO dbo.order_items (order_item_id, order_id, product_id, quantity, unit_price) VALUES (?, ?, ?, ?, ?)",
    order_items
)

conn.commit()

print("Order items inserted successfully!")
```
